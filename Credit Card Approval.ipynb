```python
pivot_tb = credit.pivot(index='ID', columns='MONTHS_BALANCE', values='STATUS')
```

```python
credit.head()
```

```python
grouped = credit.groupby('ID')
```

```python
grouped['MONTHS_BALANCE']
```

```python
pivot_tb['Open month'] = grouped['MONTHS_BALANCE'].min()
pivot_tb['End month'] = grouped['MONTHS_BALANCE'].max()
pivot_tb['ID'] = pivot_tb.index
pivot_tb = pivot_tb[['ID', 'Open month', 'End month']]
```

```python
pivot_tb.head(2)
```

```python
pivot_tb['window'] = pivot_tb['End month'] - pivot_tb['Open month'] # calculate observe window
```

```python
pivot_tb.head()
```

```python
pivot_tb.reset_index(drop = True, inplace = True)
pivot_tb.head()
```

```python
credit.head(50)
```

```python
credit1 = pd.merge(credit, pivot_tb, on = 'ID', how = 'left') # join calculated information
```

```python
credit0 = credit1.copy()
```

```python
credit0 = credit0[credit0['window'] > 20] # delete users whose observe window less than 20
```

```python
credit0['status'] = np.where((credit0['STATUS'] == '2') | (credit0['STATUS'] == '3' )| (credit0['STATUS'] == '4' )| (credit0['STATUS'] == '5'), 1, 0) # analyze > 60 days past due 
```

```python
credit0.head()
```

```python
credit0['status'] = credit0['status'].astype(np.int8) # 1: overdue 0: not
credit0['month_on_book'] = credit0['MONTHS_BALANCE'] - credit0['Open month'] # calculate month on book: how many months after opening account
credit0.sort_values(by = ['ID','month_on_book'], inplace = True)

##### denominator
denominator = pivot_tb.groupby(['Open month']).agg({'ID': ['count']}) # count how many users in every month the account was opened
denominator.reset_index(inplace = True)
denominator.columns = ['Open month','sta_sum']

##### ventage table
vintage = credit0.groupby(['Open month','month_on_book']).agg({'ID': ['count']}) 
vintage.reset_index(inplace = True)
vintage.columns = ['Open month','month_on_book','sta_sum'] 
vintage['due_count'] = np.nan
vintage = vintage[['Open month','month_on_book','due_count']] # delete aggerate column
vintage = pd.merge(vintage, denominator, on = ['Open month'], how = 'left') # join sta_sum colun to vintage table
vintage
```
